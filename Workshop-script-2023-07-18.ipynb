# Import network

```python
import pandas as pd
```

```python
colnames = ["letter_id", "date", "sender", "receiver"]
```

```python
data = pd.read_csv("sender_receiver.csv", names=colnames, header=None)
print(data)
```

```python
#G2 = nx.DiGraph()
L = nx.MultiDiGraph()
```

```python
sender = list(data["sender"])
receiver = list(data["receiver"])
dates = list(data["date"])
```

```python
for i in range(len(sender)):
    L.add_edge(sender[i], receiver[i], date=dates[i])
print(L.edges(data=True))
```

```python
L.in_degree()
```

```python
L.out_degree()
```

```python
max(nx.out_degree_centrality(L))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
```

```python
# Simple drawing

# Figure environment
f = plt.figure(1,figsize=(8,6), dpi=400)

# Network layout
pos=nx.fruchterman_reingold_layout(L)

# Drawing the whole network
nx.draw_networkx(L, pos=pos, with_labels=True, font_size=3, node_size=60, \
                 node_color='r',alpha=0.7,edge_color='gray')

# Showing the figure
plt.tight_layout()
plt.axis('off')
plt.show()
```
